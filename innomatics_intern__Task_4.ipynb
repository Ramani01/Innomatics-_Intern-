**Problem 1: Mobile Recharge Validation System**


```python
def validate_Recharge():
   valid_plans = [199,299,399,599]
   while True :
      amount = int(input("Enter the amount you want to recharge: "))
      if amount < 50 :
         print(" Invalid : Minimum recharge is Rs. 50")
      elif amount not in valid_plans :
         print("Invalid: Plan not available")
      else:
        print("Recharge Successful")
        break

        validate_Recharge()
```

** Problem 2: Inventory Reorder Alert System**


```python
def check_inventory(stock):
  for product, quantity in stock.items():
    if  quantity < 15:
      print(f"{product}: Reorder Alert")
    else:
      print(f"{product}: Stock OK")

inventory = {
    "Apples":10,
    "Bananas":25,
    "Oranges": 12
}
check_inventory(inventory)
```

**Problem 4: Cab Fare Estimator with Retry Option**


```python
def calculate_fare():
  while True:
    km = float(input("Enter the distance in kilometers: "))
    peak = input("Is it peak hour ? (yes/no): ").lower()
    fare = 50 + ( km * 12)
    if peak == "yes":
      fare *= 1.25
    print(f"Estimated Fare: Rs.{fare:.2f}")
    retry = input("Do you want to calculate another fare? (yes/no): ").lower()
    if retry != "yes":
      break

calculate_fare()
```

** Problem 5: Employee Attendance Eligibility Checker**

```python
def check_attendance (attendance):
  present_days = 0
  for day in attendance:
    if day == "P"
      present_days += 1
    percentage = (present_days /len(attendance))*100
    print(f"Attendance:{percentage}%")
    if percentage >= 75:
      return "Eligible"
    else :
      return "Not Eligible"
    status = check_attendance(["P", "A","P"])
    print(status)
```

**Problem 6: Password Strength Checker**

```python
def check_password_strength(password):
  has_digit = False
  has_special = False
  special_chars = "@#$"

  if len(password) < 8:
    return "Weak: Password should be at least 8 characters long"

  for char in password:
    if char.isdigit():
      has_digit = True
    if char in special_chars:
      has_special = True

  if has_digit and has_special:
    return "Strong"
  elif has_digit or has_special:
    return "Moderate"
  else:
    return "Weak: Does not contain a digit or special character"

pwd = input("Enter password: ")
print(check_password_strength(pwd))
```
