**Unique Words in a sentence**

```python
sentence = input("Enter a sentence:")
words = sentence.split()
unique_words = set(words)
print("Unique words count:", len(unique_words))
print("Unique words:",unique_words)
```

**Highest Salary from Employee Data**

```python
employees = {
    "Ravi":75000,
    "Anita": 68000,
    "Kiran": 72000
}
highest_employee = max(employees,key=employees.get)
print("Highest Salary:",highest_employee, "-",employees[highest_employee] )
```

**Find Maximum and Minimum Values**

```python
numbers = [45, 22,89,10,66]
maximum= numbers[0]
minimum = numbers[0]
for num in numbers:
  if num > maximum:
    maximum =num
  if num < minimum:
    minimum = num
  print("List:" ,numbers)
  print("Max:", maximum)
  print("Min:", minimum)
```

Count Products Above a Price Threshold

```python
prices = [450, 1200,899, 1500,300]
count = 0
for price in prices:
  if price > 1000:
    count +=1
    print("Products above 1000:",count)
```

**Calculate Attendance Percentage**

```python
attendance = ["P","P","A","P","P"]
present_count = attendance.count("P")
total_days =len(attendance)
percent = (present_count/total_days)*100
print("Attendance Percentage: ",percent)
```

**Remove Duplicate Phone Numbers**

**Count Character Frequency**

```python
string ="python"
char_count= {}
for char in string:
   if char in char_count:
      char_count[char] +=1
   else:
      char_count[char] = 1
print(char_count)
```
